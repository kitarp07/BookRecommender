```python
import matplotlib.pyplot as plt
user_id_counts = r['user_id'].nunique()

# Count of unique book_ids
book_id_counts = r['book_id'].nunique()

```

```python
# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
# Set the x locations for the bars
x = ['Unique Users : {}'.format(user_id_counts), 'Unique Books: {} '.format(book_id_counts)]

# Set the data for the bars
counts = [user_id_counts, book_id_counts]

# Plot the bars
ax.bar(x, counts,width=0.2)

# Set the y-axis label
ax.set_ylabel('Count')
# Set the title
ax.set_title('Unique User Counts vs Unique Book Counts (before preprocessing)')
# Display the plot
plt.show()

```
